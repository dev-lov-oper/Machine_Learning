```python
def detect_outlier(data):
  outlier_list=[]
  data=sorted(data)

  q1=np.percentile(data,25)
  q3=np.percentile(data,75)

  IQR=q3-q1

  lwr_bnd=q1-(1.5*IQR)
  upr_bnd=q3+(1.5*IQR)
  
  for i in data :
    if (i<lwr_bnd or i>upr_bnd):
       outlier_list.append(i)

  return outlier_list
```

```python
detect_outlier(df["Age"])
```
